```python
# 绘制 log(x+1)*20 的图像
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 1000, 100)
y = np.log(x + 1) * 14

plt.plot(x, y)
plt.show()
print(np.log(1000 + 1) * 14)
```
